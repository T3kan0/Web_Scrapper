```python
import pandas as pd
import datetime as dt
import itertools
import os
import requests
from tqdm import tqdm
from urllib.parse import urlparse, urlunparse
```

```python
df = pd.read_excel('2024ASTEPTutorApplicationFormEducation-5038934.xlsx')
df.columns
```

```python

students = df['8: UFS Student Number: '].unique()

for i, j, k, l in zip(students, df['28: Please attach your CV'], df['29: Please attach your full academic record'], df['30: Please attach certified copy of ID. ']):
    os.mkdir(str(i)) # creating std number directory.
    os.chdir(str(i)) # moving into the created directory.
    # Check if the URL is valid and add a default scheme (http://) if missing
    parsed_url = urlparse(j)
    if not parsed_url.scheme:
        url = urlunparse(("http",) + parsed_url[1:])
    
    try:   
        cv_req = requests.get(j, stream = True) # request CV url 
        ac_req = requests.get(k, stream = True) # request Academic Records url 
        id_req = requests.get(l, stream = True) # request ID url 
        if cv_req.status_code == 200:
            # Define file sizes to be used in the progress bar
            cv_size = int(cv_req.headers.get('content-length', 0))
            ac_size = int(ac_req.headers.get('content-length', 0))
            id_size = int(id_req.headers.get('content-length', 0))
            total_size = cv_size + ac_size + id_size
            # Use tqdm to create the progress bar
            with tqdm(total=total_size, unit='B', unit_scale=True) as pbar:
                with open('Curriculum_Vitae.pdf', 'wb') as cv: # saving CV
                    for chunk in cv_req.iter_content(chunk_size=1024):
                        if chunk:
                            cv.write(chunk)
                            pbar.update(len(chunk))
                with open('Academic_Records.pdf', 'wb') as ac: # saving AR
                    for chunk in ac_req.iter_content(chunk_size=1024):
                        if chunk:
                            ac.write(chunk)
                            pbar.update(len(chunk))
                with open('Identity_Book.pdf', 'wb') as Id: # saving ID
                    for chunk in id_req.iter_content(chunk_size=1024):
                        if chunk:
                            Id.write(chunk)
                            pbar.update(len(chunk))
         
        else:
            print(f"Error: Unable to access the URL. Status code: {cv_req.status_code}")
    except requests.exceptions.MissingSchema as e:
        print(f"Error: {e}")
    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
    
    os.chdir('..')
```
